```python
import pandas as pd
import numpy as np
```

```python
#해외 진출 선수 데이터 
total_player_list = pd.read_csv('../kwon/해외진출선수.csv', low_memory=False)
total_player_list.head()
```

```python
#골든 글러브 수상자 데이터
golden_glove = pd.read_csv('Golden Glove list.csv', low_memory=False)
golden_glove.head()
```

```python
#각 선수의 골든 글러브 수상 횟수 count
cnt_golden_glove = pd.Series()

for column in golden_glove.columns[1:]:
    # 공백 앞의 첫 번째 단어 추출
    words = golden_glove[column].str.split(' ').str[0]
    # 단어를 count
    cnt_series = words.value_counts()
    
    # 기존 Series에 단어 count 추가
    cnt_golden_glove = cnt_golden_glove.add(cnt_series, fill_value=0)

# Series를 데이터프레임으로 변환
cnt_golden_glove = cnt_golden_glove.reset_index()
cnt_golden_glove.columns = ['Player', 'Count']

print(cnt_golden_glove)
```

```python
# '이름' 열과 'Player' 열을 기준으로 'Count' 값을 매핑하여 'cnt_golden' 열을 추가
total_player_list['cnt_golden'] = total_player_list['이름'].map(cnt_golden_glove.set_index('Player')['Count'])

# 새로운 데이터프레임 생성
total_player_list_golden = total_player_list

# 결과 출력
print(total_player_list_golden)
```

```python
#골든글러브 횟수가 보이지 않아 확인
print(total_player_list_golden.iloc[:, 8:])
```

```python
# MVP 수상 데이터
mvp_list = pd.read_csv('MVP LIST.csv', low_memory=False)
mvp_list.head()
```

```python
#각 선수의 MVP 수상 횟수 count
cnt_mvp = pd.Series()

for column in mvp_list.columns[1:]:
    # 공백 앞의 첫 번째 단어 추출
    mvp_words = mvp_list[column].str.split(' ').str[0]
    # 단어를 count
    cnt_mvp_words = mvp_words.value_counts()
    
    # 기존 Series에 단어 count 추가
    cnt_mvp = cnt_mvp.add(cnt_mvp_words, fill_value=0)

# Series를 데이터프레임으로 변환
cnt_mvp = cnt_mvp.reset_index()
cnt_mvp.columns = ['Player', 'Count']

print(cnt_mvp)
```

```python
# '이름' 열과 'Player' 열을 기준으로 'Count' 값을 매핑하여 'cnt_mvp' 열을 추가
total_player_list['cnt_mvp'] = total_player_list['이름'].map(cnt_mvp.set_index('Player')['Count'])

# 새로운 데이터프레임 생성
total_player_list_golden_mvp = total_player_list_golden

# 결과 출력
print(total_player_list_golden_mvp.iloc[:, 8:])
```

```python
#올스타전 출전 현황 데이터
all_star = pd.read_csv('../kwon/올스타전출전현황.csv', low_memory=False)
all_star.head()
```

```python
# 'Unnamed: 0' 열을 'player' 열로 재명명
all_star.rename(columns={'Unnamed: 0': 'player'}, inplace=True)
all_star.rename(columns={'합계': 'cnt_all_star'}, inplace=True)

all_star_cnt = all_star[['player', 'cnt_all_star']]

# 결과 출력
print(all_star_cnt)
```

```python
# 병합(Merge)을 통해 데이터프레임 결합
player_list_not_article = pd.merge(total_player_list_golden_mvp, all_star_cnt, left_on='이름', right_on='player', how='left')


# 'player' 열을 제거
player_list_not_article.drop(columns=['player'], inplace=True)

# 결과 출력
print(player_list_not_article.iloc[:, 10:])
```

```python
#기사언급량 데이터 
article = pd.read_csv('../kwon/복귀선수기사량.csv', low_memory=False)
article.head()
```

```python
#이름열 삭제 우려로 열 이름 변경
article.rename(columns={'이름': 'player'}, inplace=True)
article_cnt = article[['player', '기사량']]

# 결과 출력
print(article_cnt)
```

```python
# 병합(Merge)을 통해 데이터프레임 결합
player_list = pd.merge(player_list_not_article, article_cnt, left_on='이름', right_on='player', how='left')


# '이름' 열을 제거
player_list.drop(columns=['player'], inplace=True)

# 결과 출력
print(player_list.iloc[:, 10:])
```

```python
# CSV 파일로 저장
player_list.to_csv('player_list.csv', index=False, encoding='utf-8-sig')
```
